```python

%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime as dt

df1 = pd.read_csv('https://www.mhlw.go.jp/content/pcr_positive_daily.csv')
df2 = pd.read_csv('https://www.mhlw.go.jp/content/pcr_tested_daily.csv')

df3 = pd.merge(df1, df2, on="日付", how='outer')
df3.to_csv('covid_data/pcr_daily.csv')

df3 = df3.rename(columns={"PCR 検査陽性者数(単日)": "pcr_positive_daily",
                          "PCR 検査実施件数(単日)": "pcr_tested_daily", 
                          "日付": "day"})

df3 = df3.set_index("day")
#print(df3)

df3.plot(figsize=(16,4))
plt.title('pcr_daily')
plt.show()

ax = df3.plot(figsize=(16,4))
fig = ax.get_figure()
fig.savefig('covid_img/pcr_positive_tested_daily.png')
```
